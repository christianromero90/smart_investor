```python
def smart_investor():  

    # Line 4 to 159 creating the Risk Profile survey (taken from Schwab: shorturl.at/uwz59 )  
    risk = 0 # Risk Tolerance score
    score = 0 # Time Horizon score
    name = str(input("Please Enter your name: "))

    print('\n')
    print('I plan to begin withdrawing money from my investments in? ')
    print('\n')
    print("A: Less than 3 years")
    print("B: 3–5 years")
    print("C: 6–10 years")
    print('D: 11 years or more')

    # Each QX is attributed a score for the final count 
    Q1 = str(input())

    if Q1== 'A' or Q1== 'a':
        score = score + 1
    elif Q1 == 'B' or Q1== 'b':
        score = score + 3
    elif Q1 == 'C'or Q1 == 'c':
        score = score + 7
    else:
        score = score + 10

    print('\n')  
    print('__'*58)
    print("Once I begin withdrawing funds from my investments, I plan to spend all of the funds in?")
    print('\n')
    print("A: Less than 2 years")
    print("B: 2–5 years")
    print("C: 6–10 years")
    print('D: 11 years or more')


    Q2 = str(input())

    if Q2== 'A' or Q2== 'a':
        score = score + 0
    elif Q2  == 'B' or Q2== 'b':
        score = score + 1
    elif Q2 == 'C' or Q2== 'c':
        score = score + 4
    else:
        score = score + 8



    print('\n')  
    print('__'*58)
    print("I would describe my knowledge of investments as?")
    print('\n')
    print("A: None")
    print("B: Limited")
    print("C: Good")
    print('D: Extensive')

    Q3 = str(input())

    if Q3== 'A' or Q3== 'a':
        risk = risk + 1
    elif Q3  == 'B' or Q3== 'b':
        risk = risk + 3
    elif Q3 == 'C' or Q3== 'c':
        risk = risk + 7
    else:
        risk = risk + 10



    print('\n')  
    print('__'*58)    
    print("When I invest my money, I am: ")
    print('\n')
    print("A: Most concerned about my investment losing value")
    print("B: Equally concerned about my investment losing or gaining value")
    print('C: Most concerned about my investment gaining value')

    Q4 = str(input())

    if Q4== 'A' or Q4== 'a':
        risk = risk + 0
    elif Q4  == 'B' or Q4== 'b':
        risk = risk + 4
    else:
        risk = risk + 8



    print('\n')  
    print('__'*58)    
    print("Select the investments you would prefer to own: ")
    print('\n')
    print("A: Bonds and/or Bond funds")
    print("B: Stocks and/or Stock funds")
    print('C: International securities and/or International funds')

    Q5 = str(input())

    if Q5== 'A' or Q5== 'a':
        risk = risk + 3
    elif Q5  == 'B' or Q5== 'b':
        risk = risk + 6
    else:
        risk = risk + 8


    print('\n')  
    print('__'*58)
    print("Consider this scenario: ")
    print('Imagine that in the past three months, the overall stock market lost 25% of its value. An individual stock investment you own also lost 25% of its value. What would you do?')
    print('\n')
    print("A: Sell all my shares")
    print("B: Sell some of my shares")
    print('C: Do Nothing')
    print('D: Buy more shares')

    Q6 = str(input())

    if Q6== 'A' or Q6== 'a':
        risk = risk + 0
    elif Q6  == 'B' or Q6== 'b':
        risk = risk + 2
    elif Q6 == 'C' or Q6== 'c':
        risk = risk + 5
    else:
        risk = risk + 8

    # Line 134 to 139 creating a table of hypothetical investments from which the user will have to choose one    
    Plan = ['A','B','C','D','E']
    Average_Annual_Return = [7.2, 9.0, 10.4, 11.7, 12.5]
    Best_Case = [ 16.3, 25.0, 33.6, 42.8, 50]
    Worst_Case = [-5.6, -12.1, -18.2, -24.0, -28.2 ]
    dataframe = pd.DataFrame({'Plan': Plan,'Average Annual Return (%)':Average_Annual_Return, 'Best Case (%)':Best_Case, 'Worst Case (%)': Worst_Case})
    dataframe.set_index('Plan', inplace=True)    
    print('\n')  
    print('__'*58)
    print("We've outlined the most likely best-case and worst-case annual returns of 5 hypothetical investment plans. Which range of possible outcomes is most acceptable to you? ")
    print('\n')
    print('The figures are hypothetical and do not represent the performance of any particular investment')
    print('\n')
    print(dataframe)

    Q7 = str(input())

    if Q7== 'A' or Q7== 'a':
        risk = risk + 0
    elif Q7  == 'B' or Q7== 'b':
        risk = risk + 3
    elif Q7 == 'C' or Q7== 'c':
        risk = risk + 6
    elif Q7 == 'D' or Q7== 'd':
        risk = risk + 8
    else:
        risk = risk + 10

    print('\n')  
    print('__'*58)
    print('How much money do you have available to invest?') # Asking how much money the user want to invest in order to show the future possible outcomes
    investment = float(input())    

    print('\n')  
    print('__'*58)

    a = 'Conservative'
    b = 'Moderately Conservative'
    c = 'Moderate'
    d = 'Moderately Aggressive'
    e = 'Aggressive'

    # Line 177 to 371 defining the score for each profile and ploting in a pie the according allocation of bonds and stock

    # Define function to plot a pie chart
    def pie_plot(bond, stock):
        colors = ['lightskyblue','lightcoral']
        explode = (0.1,0.1)  # explode 1st slice

        #Plot
        pie = {'data' : [{'type' : 'pie',
                      'name' : "Asset Allocation",
                     'labels' : ['Bond', 'Stock'],
                     'values' : [bond, stock],
                     'direction' : 'clockwise',
                     'marker' : {'colors' : ["rgb(173, 216, 230)", "rgb(255,0,0)"]}}],
              'layout' : {'title' : 'Asset Allocation'}}
        pyo.iplot(pie)


    # Conservative
    if score in range(0,5) and risk in range(0,19):
        risk_profile = a

    if score in range(5,7) and risk in range(0,16):
        risk_profile = a

    if score in range(7,10) and risk in range(0,13):
        risk_profile = a

    if score in range(10,13) and risk in range(0,12):
        risk_profile = a

    if score in range(14,19) and risk in range(0,11):
        risk_profile = a


    # Moderalty Conservative 
    if score in range(0,5) and risk in range(19,32):
        risk_profile = b

    if score in range(5,7) and risk in range(16,25):
        risk_profile = b

    if score in range(7,10) and risk in range(13,21):
        risk_profile = b

    if score in range(10,13) and risk in range(12,19):
        risk_profile = b

    if score in range(14,19) and risk in range(11,18):
        risk_profile = b



    # Moderate 
    if score in range(0,5) and risk in range(32,45):
        risk_profile = c

    if score in range(5,7) and risk in range(25,36):
        risk_profile = c

    if score in range(7,10) and risk in range(21,29):
        risk_profile = c

    if score in range(10,13) and risk in range(19,27):
        risk_profile = c

    if score in range(14,19) and risk in range(18,25):
        risk_profile = c


    # Moderately Aggressive    
    if score in range(5,7) and risk in range(36,45):
        risk_profile = d

    if score in range(7,10) and risk in range(29,38):
        risk_profile = d

    if score in range(10,13) and risk in range(27,35):
        risk_profile = d

    if score in range(14,19) and risk in range(25,32):
        risk_profile = d



    # Aggressive    
    if score in range(7,10) and risk in range(38,45):
        risk_profile = e

    if score in range(10,13) and risk in range(35,45):
        risk_profile = e

    if score in range(14,19) and risk in range(32,45):
        risk_profile = e



    if risk_profile == a:
        print(name, ' your investment profile is ', a)
        bond = 0.90
        stock = 0.10
        pie_plot(bond, stock)

    if risk_profile == b:
        print(name, ' your investment profile is ', b)
        bond = 0.65
        stock = 0.35
        pie_plot(bond, stock)

    if risk_profile == c:
        print(name, ' your investment profile is ', c)
        bond = 0.50
        stock = 0.50
        pie_plot(bond, stock)

    if risk_profile == d:
        print(name, ' your investment profile is ', d)
        bond = 0.35
        stock = 0.65
        pie_plot(bond, stock)

    if risk_profile == e:
        print(name, ' your investment profile is ', e)
        bond = 0.10
        stock = 0.90
        pie_plot(bond, stock)
    
    
    # We now ask for the current holdng of the user
    
    nbrs_holdings = int(input('How many holdings do you have ? ')) 
    acquisition_date = [] #Creating an empty list to create later a dataframe
    ticker = []
    qty = []
    unit_cost = []
    cost_basis = []

    for i in range(nbrs_holdings): #adding element to each list
        print('\n')

        tickers_loop = input('Enter ticker symbol: ')

        acq_date = pd.to_datetime(input('Enter acquisition date (YYYY-MM-DD): '))

        quantity = float(input('Enter quantity: '))

        buy_price = round(data.DataReader(tickers_loop, 'yahoo', start = acq_date, end = acq_date+datetime.timedelta(3))['Adj Close'][0],2)

        cost__basis = quantity * buy_price

        ticker.append(tickers_loop) # Add the ticker input by the user to the list "Ticker" of line 376

        acquisition_date.append(acq_date)
        qty.append(quantity)
        unit_cost.append(buy_price)
        cost_basis.append(cost__basis)
        

        
    # Creating the Dataframe
    holdings_df = pd.DataFrame({'Acquisition Date': acquisition_date,'Ticker':ticker, 'Quantity':qty, 'Unit Cost': unit_cost, 'Cost Basis':cost_basis})
    holdings_df.set_index('Ticker', inplace=True)

    start_date = "2015-01-01"
    today = (datetime.datetime.today()- datetime.timedelta(1)).strftime('%Y-%m-%d')
    tickers = holdings_df.index.tolist()
    
    

    # Define function to get just the Adjusted Close price
    def get_close(tickers, start, end):
        stock_data = data.DataReader(tickers, data_source='yahoo', start = start, end = end)['Adj Close']
        return stock_data
    
    
    

    # Getting the stocks of interest returns to use in CAPM calculation
    stock_adj_close = get_close(tickers, start_date, today)
    stock_adj_close.reset_index(inplace=True)
    stock_returns = stock_adj_close.iloc[:, 1:].pct_change().dropna()


    # Retrieving the risk free rate (daily) to use in CAPM calculation
    rfr = (get_close('^tyx', start_date, today).mean())/100 #defining the risk free rate
    rfr_daily = rfr/252

    # Using a Bond ETF as our low risk asset
    bond_etf = get_close(['bnd'], start_date, today)
    bond_etf = bond_etf.reset_index().drop('Date', axis=1)
    bond_etf_returns = bond_etf.pct_change().dropna()

    #Initializing some empty lists to fill in the loop
    holdings_df['todays holdings'] = 0
    holdings_df['todays weights'] = 0
    holdings_df['expected return'] = 0
    holdings_df['standard deviation'] = 0

    # Saving EXPECTED RETURN and Standard Deviation of the stocks of interest in a dataframe
    expected_returns_df = pd.DataFrame(stock_returns.mean().tolist(), columns = ['Exp Ret'], index=tickers)
    standard_deviation_df = pd.DataFrame(stock_returns.std().tolist(), columns = ['Std Dev'], index=tickers)

    
    
    # Filling holdings_df 
    for x in tickers:

        todays_adj_close = stock_adj_close[x].iloc[-1]
        holdings_df.loc[x, 'todays holdings'] = todays_adj_close * holdings_df.loc[x,'Quantity']

        holdings_df.loc[x, 'expected return'] = expected_returns_df.loc[x, 'Exp Ret'] * 252

        holdings_df.loc[x, 'standard deviation'] = standard_deviation_df.loc[x, 'Std Dev'] * np.sqrt(252)

    suma = holdings_df['todays holdings'].sum()
    
    
    
    for x in tickers: 
        holdings_df.loc[x, 'todays weights'] = holdings_df.loc[x, 'todays holdings']/suma

    
    # Creating pie charts to show user's holdings today
    pie2 = {'data' : [{'type' : 'pie',
                      'name' : "Initial weights",
                     'labels' : tickers,
                     'values' : holdings_df['todays weights'].tolist(),
                     'direction' : 'clockwise',}],
          'layout' : {'title' : 'Initial weights'}}
    pyo.iplot(pie2)  
    
    

    # Line 407 to 435 Defining function to maximize sharpe ratio
    def max_sharpe(holdings, returns):

        np.random.seed(0)
        num_portfolios = 6000
        all_weights = np.zeros((num_portfolios,len(holdings)))
        return_array = np.zeros(num_portfolios)
        vol_array = np.zeros(num_portfolios)
        sharpe_array = np.zeros(num_portfolios)

        for x in range(num_portfolios):

            weights = np.array(np.random.random(len(holdings)))
            weights = weights/np.sum(weights)

            all_weights[x,:] = weights

            return_array[x] = np.sum((returns.mean() * weights * 252))

            vol_array[x] = np.sqrt(np.dot(weights.T, np.dot(returns.cov()*252, weights)))

            sharpe_array[x] = (return_array[x]-rfr)/vol_array[x]

        max_sharpe_weights = all_weights[sharpe_array.argmax(),:]
        max_sharpe_weights_df = pd.DataFrame(max_sharpe_weights, columns = ['weights'], index = returns.columns.tolist())

        max_sharpe_return = return_array[sharpe_array.argmax()]
        max_sharpe_vol = vol_array[sharpe_array.argmax()]

        return max_sharpe_return, max_sharpe_vol, max_sharpe_weights_df, vol_array, return_array, sharpe_array

    
    # Save function returns
    function_returns = max_sharpe(holdings_df, stock_returns)

    
    # Visualization Miu-Sigma sharpe Ratio and Optimal weights
    fig1 = go.Figure()

    fig1.add_trace(
        go.Scatter(x= function_returns[3], 
                   y= function_returns[4], 
                   mode='markers',
                   text = function_returns[5],
                   marker = dict(color=function_returns[5],
                                     colorscale='Viridis',
                                     showscale=True),
                   name = 'Sharpe Ratio'
                  ))

    fig1.add_trace(
        go.Scatter(x= [function_returns[1]], 
                   y= [function_returns[0]],
                   mode='markers',
                   text = (function_returns[0]-rfr)/function_returns[1],
                   marker = dict(size = 10,
                                    symbol = 'pentagon',
                                    color = 'rgb(255,0,0)'),
                   name = 'Max Sharpe Ratio'
                   ))

    fig1.show()

    #pie with current holdings optimized
    pie3 = {'data' : [{'type' : 'pie',
                      'name' : "Optimized weights",
                     'labels' : function_returns[2].index.tolist(),
                     'values' : function_returns[2]['weights'],
                     'direction' : 'clockwise',}],
          'layout' : {'title' : 'Optimized weights'}}
    pyo.iplot(pie3) 

    
    
    # Loop to tell user how the weights of the assets should change
    for x in tickers:
        print('The weight of: ', x , ' Should change from ',
              round(holdings_df.loc[x, 'todays weights']*100,3),'%',
              'to ', round(function_returns[2].loc[x,'weights']*100, 3),'%')
        
    
    # Create a copy of the holdings_df dataframe that will not modify the original
    holdings_df_new = holdings_df.copy(deep=True)
    holdings_df_new['new weights'] = 0
    removed_tickers = []

    # Loop to remove tickers whose weights is lower than 2% 
    for x in tickers:
        if function_returns[2].loc[x, 'weights'] >= 0.02: 
            holdings_df_new.loc[x, 'new weights'] = function_returns[2].loc[x,'weights']
        else:
            removed_tickers.append(x),
            holdings_df_new.drop(index=x, inplace=True)

            
    # Remove tickers below 2% from 'returns' and reset index to numerical, drop the date column
    rem_stock_returns = stock_returns.drop(removed_tickers, axis=1)


    #Re-calculate returns after dropping stocks with low weights
    return_array = np.zeros(len(rem_stock_returns))

    new_weights = list(holdings_df_new['new weights'])

    for i in rem_stock_returns.index.tolist():
        return_array = rem_stock_returns.dot(new_weights)

    return_array_df = pd.DataFrame(return_array).rename(columns = {0:'Returns'})
    
    
    
    # Ask user which stocks he would like to add to the portfolio
    print('\n')
    new_stocks = input("Which stocks would you like to add to your portfolio? (Please write ticker symbol): ").split()

    new_stock_prices = get_close(new_stocks, start_date, today)
    new_stock_returns = new_stock_prices.pct_change().dropna()
    new_stock_returns.reset_index(inplace=True)
    new_stock_returns.drop('Date', axis=1, inplace=True)
    
    
    # Define function to calculate Beta for Modern Portoflio Theory Optm.
    def get_beta(returnsportfolio, returns_new_stock):
        covariance = returnsportfolio.cov(returns_new_stock)

        variance = returnsportfolio.var()

        beta = covariance/variance

        return beta

    
    
    # Applying the portfolio improment rule
    for x in new_stocks:

        beta = get_beta(return_array_df['Returns'], new_stock_returns[x])

        if new_stock_returns[x].mean() > (rfr_daily + (beta * (return_array_df['Returns'].mean()-rfr_daily))):

            # Add New stock to holding_df_new 
            holdings_df_new = holdings_df_new.reset_index().append({'Ticker':x, 'Acquisition Date':today}, ignore_index=True)

            # Add new stock returns to 'returns' dataframe
            rem_stock_returns = pd.merge(rem_stock_returns, new_stock_returns[x], left_index=True, right_index=True)

            # Calculate max Sharpe for this new portfolio
            function_returns_1 = max_sharpe(holdings_df_new, rem_stock_returns)
            max_sharpe_weights = function_returns_1[2]

            #Re-calculate return array
            for i in rem_stock_returns.index.tolist():
                return_array = rem_stock_returns.dot(list(max_sharpe_weights['weights']))

            return_array_df = pd.DataFrame(return_array).rename(columns = {0:'Returns'})
            
    holdings_df_new = holdings_df_new.drop('index', axis=1)
    holdings_df_new.set_index('Ticker', inplace=True)

    
    suma = suma + investment

    # Updating Df with new stock info
    for x in holdings_df_new.index.tolist():
        holdings_df_new.loc[x,'new weights'] = max_sharpe_weights.loc[x,'weights']
        holdings_df_new.loc[x,'todays holdings'] = holdings_df_new.loc[x,'new weights'] * suma


    # Expected returns and volatility of the portfolio (stock + bond) (daily)
    asset_allocation_returns = (bond * bond_etf_returns['bnd']) + (stock * return_array)
    asset_allocation_mean = asset_allocation_returns.mean()
    asset_allocation_std = asset_allocation_returns.std()


    # Pie Chart showing the optimized weights
    pie4 = {'data' : [{'type' : 'pie',
                      'name' : "Optimized weights",
                     'labels' : holdings_df_new.index.tolist(),
                     'values' : holdings_df_new['new weights'].tolist(),
                     'direction' : 'clockwise',}],
          'layout' : {'title' : 'Optimized weights'}}
    pyo.iplot(pie4) 


    # Will display the Max Sharpe Ratio of the portfolio in mu-sigma space
    fig2 = go.Figure() 

    fig2.add_trace(
        go.Scatter(x= function_returns_1[3] , 
                   y= function_returns_1[4], 
                   mode='markers', 
                   text = function_returns_1[5],
                   marker = dict(color=function_returns_1[5],
                                     colorscale='Viridis',
                                     showscale=True),
                   name = 'Sharpe Ratio'
                  ))

    fig2.add_trace(
        go.Scatter(x= [function_returns_1[1]] , 
                       y= [function_returns_1[0]],
                       mode='markers',
                       text = [(function_returns_1[0]-rfr)/function_returns_1[1]],
                       marker = dict(size = 10,
                                    symbol = 'pentagon',
                                    color = 'rgb(255,0,0)'),
                       name = 'Max Sharpe Ratio'
        ))

    fig2.show() 

    # Fill the values of expected return and standard deviation for the new stocks added to 'holdings_df_new'
    for a in holdings_df_new.index.tolist():
        holdings_df_new.loc[a,'expected return'] = 252*(get_close(a, start_date, today).pct_change().mean())
        holdings_df_new.loc[a,'standard deviation'] = np.sqrt(252) * (get_close(a, start_date, today).pct_change().std())


    # Visually compare the Max Sharpe of the portfolio vs the individual sharpe ratio of the stocks
    fig3 = go.Figure() 

    fig3.add_trace(
        go.Scatter(x = holdings_df_new['standard deviation'] , 
                   y = holdings_df_new['expected return'],
                           mode = 'markers',
                           text = np.asarray(holdings_df_new.index.tolist()),
                           marker = dict(size = 10,
                                         color=holdings_df_new['new weights']
                                    ),
                           name = 'Other Stocks'
                  ))

    fig3.add_trace(
        go.Scatter(x= [function_returns_1[1]] , 
                       y= [function_returns_1[0]],
                       mode='markers',
                       text = [(function_returns_1[0]-rfr)/function_returns_1[1]],
                       marker = dict(size = 10,
                                    symbol = 'pentagon',
                                    color = 'rgb(255,0,0)'),
                       name = 'Max Sharpe Ratio'
        ))

    fig3.show()



    # Creating a Monte Carlo Simulation with a VaR
    random.seed(100)
    darts = 100
    days = 252
    zero_data = np.zeros(shape=(darts, days))
    monte_carlo = pd.DataFrame(zero_data)


    for a in monte_carlo.index.tolist():
        for p in monte_carlo.columns.tolist():
            b = (np.random.normal(asset_allocation_mean, asset_allocation_std))
            monte_carlo.loc[a,p] = (b + 1)

    monte_carlo_new = monte_carlo.copy(deep=True) 

    for f in monte_carlo_new.index.tolist():
        monte_carlo_new.loc[f,:] = monte_carlo_new.loc[f,:].cumprod(axis=0)

    monte_carlo_new.sort_values(by=251, inplace=True)

    monte_carlo_new = monte_carlo_new.reset_index().drop('index', axis=1)

    zero_data_2 = np.zeros(shape=(int(darts/10), days))
    averages = pd.DataFrame(zero_data_2)
    #display(averages)

    for f in averages.index.tolist():
        begin = f*10
        end = begin + 10
        df = monte_carlo_new.iloc[begin:end, :].mean()
        averages.loc[f] = df

    averages_trans = averages.T



    linegraph = [go.Scatter(x = np.asarray(averages_trans.index.tolist()) ,
                            y = averages_trans.iloc[:,i],
                            mode = 'lines') for i in averages_trans.columns.tolist()]

    linelayout = go.Layout(title = 'Simulated cummulative returns (rebased)',
                       xaxis = {'title': 'days'},
                       yaxis = {'title': 'returns (%)'},
                       hovermode = 'closest')

    fig5 = go.Figure(data = linegraph, layout=linelayout)
    fig5.update_layout(showlegend=False)

    pyo.iplot(fig5) # Graph Monte Carlo Simulation

    print('\n')
    print('__'*58)
    print('\n')
    # Asking if the user want to reuse the Robo and relaunch the simulation
    redo = input("Would you like to redo a simualtion?(Please type: 'Y' for yes and 'N' for no)")
    if redo == 'Y' or redo == 'y':
        smart_investor()
    elif redo == 'N' or redo == 'n':
        print('Thank you for using SmartInvestor')
```
